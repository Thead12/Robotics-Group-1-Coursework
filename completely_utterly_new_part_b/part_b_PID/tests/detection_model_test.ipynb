```python
from ultralytics import YOLO
```

```python
# initialise YOLO models
full_model = YOLO("yolo11l_half.engine")
human_only_model = YOLO('best.pt')
```

```python
import camera
camera.Camera()
camera.start()
```

```python
import time

test_length = range(300)

start_time = time.time()


latest_color_frame = None
latest_depth_frame = None

for i in test_length:
    latest_color_frame, latest_depth_frame = camera.get_frame()

    results = full_model(latest_color_frame, classes=[0], verbose=False)

end_time = time.time()
elapsed_time = end_time - start_time

print("Full model inference on 300 frames completed.")
print(f"Elapsed Time: {elapsed_time} seconds")

control_frequency = len(test_length)/elapsed_time
print(f"Control frequency = ", control_frequency)
```
